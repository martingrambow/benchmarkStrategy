```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns


folder = "../../results_all/app_AA-test_vm/"
filename = "../../results_all/app_AA-test_vm/app_AA-test_vm"
print(folder)

start = 0.05
end = 0.2
```

```python
all_runs = glob.glob(folder + "*")
rows = []
for fullRunFolder in all_runs:
    all_files = glob.glob(fullRunFolder + "/*")
    size = len(all_files)

    if size != 22:
        print(f"{size} elements in folder {fullRunFolder}, skip.")
    else:
        tmp = fullRunFolder.split("\\")

        #Parse run and commit number from path structure
        run = tmp[len(tmp)-1]
        print(f"Run {run}")

        if (int(run) <= 5):

            #Read latency file for both types (old and new)
            insertsOld1 = pd.read_csv(fullRunFolder + "/latenciesInserts1Old.csv",
                             header=None,
                             delim_whitespace=True,
                             names=["batch","latency (ms)"])
            insertsOld1.sort_values(by=["batch"], inplace=True)
            number = len(insertsOld1.index)
            removeFirst = int(start * number)
            removeLast = int(end * number)
            insertsOld1 = insertsOld1.iloc[removeFirst:-removeLast , :]

            for index, row in insertsOld1.iterrows():
                rows.append({
                    "run":run,
                    "number": row["batch"],
                    "type" : "inserts",
                    "version" : "base",
                    "latency (ms)" : row["latency (ms)"]
                })

            insertsOld2 = pd.read_csv(fullRunFolder + "/latenciesInserts2Old.csv",
                             header=None,
                             delim_whitespace=True,
                             names=["batch","latency (ms)"])
            insertsOld2.sort_values(by=["batch"], inplace=True)
            number = len(insertsOld2.index)
            removeFirst = int(start * number)
            removeLast = int(end * number)
            insertsOld2 = insertsOld2.iloc[removeFirst:-removeLast , :]

            for index, row in insertsOld2.iterrows():
                rows.append({
                    "run":run,
                    "number": row["batch"],
                    "type" : "inserts",
                    "version" : "base",
                    "latency (ms)" : row["latency (ms)"]
                })

            insertsNew1 = pd.read_csv(fullRunFolder + "/latenciesInserts1New.csv",
                             header=None,
                             delim_whitespace=True,
                             names=["batch","latency (ms)"])
            insertsNew1.sort_values(by=["batch"], inplace=True)
            number = len(insertsNew1.index)
            removeFirst = int(start * number)
            removeLast = int(end * number)
            insertsNew1 = insertsNew1.iloc[removeFirst:-removeLast , :]

            for index, row in insertsNew1.iterrows():
                rows.append({
                    "run":run,
                    "number": row["batch"],
                    "type" : "inserts",
                    "version" : "variation",
                    "latency (ms)" : row["latency (ms)"]
                })

            insertsNew2 = pd.read_csv(fullRunFolder + "/latenciesInserts2New.csv",
                             header=None,
                             delim_whitespace=True,
                             names=["batch","latency (ms)"])
            insertsNew2.sort_values(by=["batch"], inplace=True)
            number = len(insertsNew2.index)
            removeFirst = int(start * number)
            removeLast = int(end * number)
            insertsNew2 = insertsNew2.iloc[removeFirst:-removeLast , :]

            for index, row in insertsNew2.iterrows():
                rows.append({
                    "run":run,
                    "number": row["batch"],
                    "type" : "inserts",
                    "version" : "variation",
                    "latency (ms)" : row["latency (ms)"]
                })

            queries1Old = pd.read_csv(fullRunFolder + "/latenciesQueries1Old.csv",
                             header=None,
                             delim_whitespace=True,
                             names=["query","latency (ms)"])
            queries1Old.sort_values(by=["query"], inplace=True)
            number = len(queries1Old.index)
            removeFirst = int(start * number)
            removeLast = int(end * number)
            queries1Old = queries1Old.iloc[removeFirst:-removeLast , :]


            for index, row in queries1Old.iterrows():
                rows.append({
                    "run":run,
                    "number": row["query"],
                    "type" : "simple queries",
                    "version" : "base",
                    "latency (ms)" : row["latency (ms)"]
                })

            queries2Old = pd.read_csv(fullRunFolder + "/latenciesQueries2Old.csv",
                             header=None,
                             delim_whitespace=True,
                             names=["query","latency (ms)"])
            queries2Old.sort_values(by=["query"], inplace=True)
            number = len(queries2Old.index)
            removeFirst = int(start * number)
            removeLast = int(end * number)
            queries2Old = queries2Old.iloc[removeFirst:-removeLast , :]

            for index, row in queries2Old.iterrows():
                rows.append({
                    "run":run,
                    "number": row["query"],
                    "type" : "group-by queries",
                    "version" : "base",
                    "latency (ms)" : row["latency (ms)"]
                })

            queries1New = pd.read_csv(fullRunFolder + "/latenciesQueries1New.csv",
                             header=None,
                             delim_whitespace=True,
                             names=["query","latency (ms)"])
            queries1New.sort_values(by=["query"], inplace=True)
            number = len(queries1New.index)
            removeFirst = int(start * number)
            removeLast = int(end * number)
            queries1New = queries1New.iloc[removeFirst:-removeLast , :]

            for index, row in queries1New.iterrows():
                rows.append({
                    "run":run,
                    "number": row["query"],
                    "type" : "simple queries",
                    "version" : "variation",
                    "latency (ms)" : row["latency (ms)"]
                })

            queries2New = pd.read_csv(fullRunFolder + "/latenciesQueries2New.csv",
                             header=None,
                             delim_whitespace=True,
                             names=["query","latency (ms)"])
            queries2New.sort_values(by=["query"], inplace=True)
            number = len(queries2New.index)
            removeFirst = int(start * number)
            removeLast = int(end * number)
            queries2New = queries2New.iloc[removeFirst:-removeLast , :]

            for index, row in queries2New.iterrows():
                rows.append({
                    "run":run,
                    "number": row["query"],
                    "type" : "group-by queries",
                    "version" : "variation",
                    "latency (ms)" : row["latency (ms)"]
                })

# Convert to data frame
df_result = pd.DataFrame(rows)
df_result.sort_values(by=["type","number","version"], inplace=True)
df_result.describe()
# Save to file
df_result.to_csv(filename + ".csv")
```
